# Applications of Fourier Transform to Derivatives and Shifting Functions

## Fourier Transform and Fourier Derivatives

Differentiation can be done term by term on the spectral domain, and can be transformed back, using inverse Fourier transform to obtain the derivative in the time domain.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000  # number of sample points
L = 2*np.pi  # length of interval
dx = L/N  # sample spacing
x = np.linspace(0, L, N, endpoint=False)
y = np.sin(x)

# Compute the Fourier coefficients
c = np.zeros_like(x, dtype=np.complex128)
for n in range(-N//2, N//2):
    for i in range(N):
        c[n+N//2] += y[i]*np.exp(-1j*n*x[i])*dx
    c[n+N//2] *= 2/L

# Compute the Fourier series
def fourier_series(c, L, n_terms, x):
    f = np.zeros_like(x)
    for i in range(len(x)):
        for n in range(-n_terms, n_terms+1):
            f[i] += c[n+N//2]*np.exp(1j*n*x[i])/2
    return f

n_terms = 50
f = fourier_series(c, L, n_terms, x)

# Compute the Fourier series of the first derivative
dfdx = np.zeros_like(f)
for i in range(len(dfdx)):
    for n in range(-n_terms, n_terms+1):
        dfdx[i] += 1j*n*c[n+N//2]*np.exp(1j*n*x[i])/2

# Compute the Fourier series of the second derivative
d2fdx2 = np.zeros_like(f)
for i in range(len(d2fdx2)):
    for n in range(-n_terms, n_terms+1):
        d2fdx2[i] -= n**2*c[n+N//2]*np.exp(1j*n*x[i])/2

# Plot the results
plt.plot(x, y, label='f(x)')
plt.plot(x, dfdx.real, label='first derivative')
plt.plot(x, d2fdx2.real, label='second derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plots of f(x) = sin(x) and its first and second derivative')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000  # number of sample points
L = 2*np.pi  # length of interval
dx = L/N  # sample spacing
x = np.linspace(0, L, N, endpoint=False)
y = np.sin(x)

# Compute the Fourier coefficients of the original function
c = np.zeros_like(x, dtype=np.complex128)
for n in range(-N//2, N//2):
    for i in range(N):
        c[n+N//2] += y[i]*np.exp(-1j*n*x[i])*dx
    c[n+N//2] *= 2/L

# Shift the Fourier coefficients by multiplying them by the phase factor
shift = -1  # shift amount
for n in range(-N//2, N//2):
    c[n+N//2] *= np.exp(1j*2*np.pi*n*shift/L)

# Compute the Fourier series of the shifted function
n_terms = 50
f_shifted = np.zeros_like(x)
for i in range(len(x)):
    for n in range(-n_terms, n_terms+1):
        f_shifted[i] += c[n+N//2]*np.exp(1j*n*x[i])/2

# Plot the original function and the shifted function
plt.plot(x, y, label='Original function')
plt.plot(x, f_shifted.real, label='Shifted function')
plt.legend()
plt.show()
```
